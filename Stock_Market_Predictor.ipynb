```python
import yfinance as yf
```

```python
#Downloading S&P 500 price data
# initialize a ticker class
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period="max")
```

```python
sp500
```

```python
sp500.index
```

```python
#plot closing data against the index of S&P price history
sp500.plot.line(y="Close", use_index=True)
```

```python
#Cleaning Data
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
#Stock price increase prediction
#Pandas shift method
sp500["Tomorrow"] = sp500["Close"].shift(-1)
sp500
```

```python
#Predict stock market target with ML
#Check Tomorrow's price of Stock with Current Date
#Convert boolean to int using astype for conversion for usage in ML
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
sp500
```

```python
#Take out irrelevant historical data
#first isolate the rows with 1990 using loc method
sp500 = sp500.loc["1990-01-01":].copy()
sp500
```

```python
#Training ML model
#random forest can pick up non-linear relationships
from sklearn.ensemble import RandomForestClassifier
```

```python
#n_estimators is the number of individual decision trees we want to train, higher accuracy
#min_samples_split, prevents overfitting of model
#random_state makes the randomization into a predictable sequence
model = RandomForestClassifier(n_estimators=100, min_samples_split=100,random_state=1)
#time series data, can't use cross validation
train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

#list of columns that will help predict the target
predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
#Measure accuracy of the model
from sklearn.metrics import precision_score
preds = model.predict(test[predictors])

```

```python
import pandas as pd
#convert predictions into pandas series so easier to read
preds = pd.Series(preds, index=test.index)
#precision_score(test["Target"], preds)

precision = precision_score(test["Target"], preds, zero_division=1)
precision
```

```python
#Plot predictions combine actual and predicted values using pandas concat method
combined = pd.concat([test["Target"], preds], axis=1) #axis=1, treat each input as a col in the dataset
combined
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined

def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()

        # Ensure the predictors are 2D arrays
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)

    return pd.concat(all_predictions)

# Perform backtest
predictions = backtest(sp500, model, predictors)
predictions["Predictions"].value_counts()
```

```python
#additional predictors of the model
#create variety of rolling averages
# calculate rolling means in past 2 days, trading week, 3 months, last year, past four years
horizons = [2, 5, 60, 250, 100]
#new list
new_predictors = []

for horizon in horizons:
  rolling_averages = sp500.rolling(horizon).mean()
  ratio_column = f"New_Ratio_{horizon}"
  sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]

  trend_column = f"Trend_{horizon}"
  #sums all the targets from past 4-5 days
  sp500[ratio_column] = sp500.shift(1).rolling(horizon).sum()["Target"]

  new_predictors += [ratio_column, trend_column]
```

```python
#GPT changes
horizons = [2, 5, 60, 250, 1000]
for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    sp500[f"Close_Ratio_{horizon}"] = sp500["Close"] / rolling_averages["Close"]
    sp500[f"Volume_Ratio_{horizon}"] = sp500["Volume"] / rolling_averages["Volume"]
    sp500[f"Trend_{horizon}"] = sp500["Target"].shift(1).rolling(horizon).sum()

sp500 = sp500.dropna()
```
